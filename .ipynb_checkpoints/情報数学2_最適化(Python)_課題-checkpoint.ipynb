```python
import pulp
```

```python
problem = pulp.LpProblem('P', pulp.LpMaximize)

#å¤‰æ•°ã‚’å®šç¾©(å¤‰æ•°å, æœ€å°å€¤, æœ€å¤§å€¤, æ•´æ•°åˆ¶ç´„)
x = pulp.LpVariable('x', 0, None)
y = pulp.LpVariable('y', 0, None)

#ç›®çš„é–¢æ•°ã‚’å®šç¾©
problem += 3*x + 4*y

#åˆ¶ç´„æ¡ä»¶ã‚’å®šç¾©
problem += 8*x + 7*y <= 50
problem += 2*x + 3*y <= 20
problem += x + 4*y <= 25

#ç¢ºèª
print(problem)
```

```python
#å•é¡Œã‚’è§£ã
status = problem.solve()
print(pulp.LpStatus[status])
print(pulp.value(problem.objective))
print(x.value())
print(y.value())
```
