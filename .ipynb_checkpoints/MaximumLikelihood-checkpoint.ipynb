```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#いびつなコイン
pr = np.arange(0, 1, 0.01)
py = pr**3 * (1-pr)**2
plt.plot(pr, py)
```

```python
import sympy as sp

p = sp.symbols('p')
lp = p**3 * (1-p)**2
sp.init_printing()
lp
```

```python
dlp = sp.diff(lp, p, 1)
dlp = sp.factor(dlp)
dlp
```

```python
solve = sp.solve(dlp, p)
solve
```

```python
#尤度関数(係数部は除く)
py1 = pr**20 * (1-pr)**80
plt.plot(pr, py1)
```

```python
#尤度関数(係数部は除く)
lp1 = p**20 * (1-p)**80
lp1
```

```python
#対数尤度関数(係数部は除く)
lp2 = sp.log(p**20 * (1-p)**80)
lp2
```

```python
#対数尤度関数を微分
dlp2 = sp.diff(lp2, p, 1)
dlp2 = sp.factor(dlp2)
dlp2
```

```python
solve1 = sp.solve(dlp2, p)
solve1
```

```python
#ポアソン分布
x1 = sp.symbols('x1')
x2 = sp.symbols('x2')
x3 = sp.symbols('x3')
mu = sp.symbols('mu')

#尤度関数
lp3 = mu**x1 / sp.factorial(x1) * sp.exp(-mu) * mu**x2 / sp.factorial(x2) * sp.exp(-mu) * mu**x3 / sp.factorial(x3) * sp.exp(-mu)
lp3

#対数尤度関数
lp4 = sp.log(lp3)
lp4
```

```python
#対数尤度関数を微分
dlp4 = sp.diff(lp4, mu, 1)
dlp4 = sp.factor(dlp4)
dlp4
```

```python
solve2 = sp.solve(dlp4, mu)
solve2 = sp.factor(solve2)
solve2
```
